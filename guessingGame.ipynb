Below is a simple number guessing game I implemented as an exercise to refamiliarize myself with basic Python syntax and controls, as it has been some time since I last worked with Python.

**Mechanism Of the Game:**
    
The game generates a number between 1 and 100, and will give a random clue to start off.  The player starts with a score of 20, losing a point for each incorrect guess.  After the first 2 guesses, the game will try to make your life easier by telling you if you're getting warmer or colder than your immediately preceding guess.  After 7 incorrect guesses, the game is programmed to feel bad for you and give you another random clue from the remaining clue pool.  

```python
import math
import random

#Give one of 3 types of hints to the player about the hidden number
def giveClue(hiddenNumber, rand):
    
    factors = []
    
    #Number of factors/IsPrime
    if rand == 0:

        #Factors of the hiddenNumber
        for i in range(1, hiddenNumber + 1):
            if hiddenNumber % i == 0:
                factors.append(i)
                
        if len(factors) > 2:
            print("Hint: the hidden number has", len(factors) - 2, "factors other than 1 and itself.")
        else:
            print("Hint: the hidden number is prime.")
    
    #IsEven/Odd
    elif rand == 1:
        if hiddenNumber % 2 == 0:
            print("Hint: the hidden number is even")
        else:
            print("Hint: the hidden number is odd")
            
    #GreaterLesser50
    else:
        if hiddenNumber > 50:
            print("Hint: the hidden number is greater than 50")
        else:
            print("Hint: the hidden number is less than 50")

#main game function
def guessingGame():
    print("The following is a number guessing game for a value between 1 and 100.\n")
    
    #generate the hidden number to guess, and an index for the randomized type of clue to give
    hiddenNumber = random.randrange(1, 101)
    clueNumber = random.randrange(0,3)
    #print(hiddenNumber)
    
    #generate a random clue
    giveClue(hiddenNumber, clueNumber)
    
    playerScore = 20  #running count of player's score
    previousGuess = 0 #used to determine whether the guess is warmer or colder
    previousDistance = 0
    currentGuess = int(input("\nEnter your first guess: ")) #initial guess handler
    
    #game loop
    while currentGuess != hiddenNumber:
        print("Guess again!\n")
        playerScore = playerScore - 1 #deduct a point for each wrong guess
        
        #give a new (still random) clue if the player is down to 13 points
        if playerScore == 13:
            rand = random.randrange(0,3)
            while rand == clueNumber:
                rand = random.randrange(0,3)
                
            giveClue (hiddenNumber, rand)
        
        currentGuess = int(input("Enter a guess: "))
        
        #check if new guess is warmer or colder than old guess
        currentDistance = abs(hiddenNumber - currentGuess)
        if previousGuess != 0 & currentGuess != hiddenNumber:
            if currentDistance < previousDistance:
                print("Warmer than your previous guess!\n ")
            else:
                print("Colder than your previous guess!\n ")
        
        #update previous guess for next iteration
        previousGuess = currentGuess
        previousDistance = currentDistance
    
    #victory
    print("You got it! The hidden number was", hiddenNumber, "and you finished with a score of", playerScore, "!")
    

random.seed()
guessingGame()
```
